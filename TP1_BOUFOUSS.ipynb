5. Implémenter une fonction initialisation (X) des paramètres d'un modèle (W et
b) de régression linéaire

```python
def initialisation (X):
    n_features = X.shape[1]  # Nombre de caractéristiques
    W = np.ones((n_features, 1))  # Initialiser W (2,1) avec des 1
    b = 0  # Initialiser b à 0
    return W, b
```

6. Implémenter une fonction model(X, W, b) des paramètres d'un modèle de
régression linéaire (formules (1) et (2))

```python
#formules (1)
def model(X, W, b):
    z = np.dot(X, W) + b
    return z
```

```python
#formules (1)
def sigmoid(z):
    A=  1 / (1 + np.exp(-z))
    return A
```

7. Implémenter une fonction log_loss(A, y) (formule (3))

```python
def log_loss(A, y):
    m = A.shape[0]
    loss = - (1 / m) * np.sum(A * np.log(y) + (1 - A) * np.log(1 - y))
    return loss
```

8. Implémenter une fonction gradients(A, X, y) (formule (4))

```python
def Gradient(A, X, y):
    m = X.shape[0]  # Nombre d'exemples
    dW = (1/m) * np.dot(X.T, (A - y))  # Dérivée par rapport aux poids
    db = (1/m) * np.sum(A - y)  # Dérivée par rapport au biais
    return dW, db
```

9. Implémenter une fonction update(dW, db, W, b, learning_rate) avec
learning_rate est le taux d'apprentissage

```python
def update(dW, db, W, b, learning_rate):
    # Mise à jour des poids
    W = W - learning_rate * dW
    # Mise à jour des biais
    b = b - learning_rate * db
    return W, b
```

10. Implémenter une fonction predict(X, W, b) avec un seuil >= 0.5

```python
def predict(X, W, b, threshold=0.5):
    # Calculer la sortie de la fonction linéaire
    z=model(X, W, b)
    
    # Appliquer la fonction sigmoïde pour obtenir une probabilité
    prob = sigmoid(z)
    
    # Appliquer le seuil pour prédire 0 ou 1
    predictions = (prob >= threshold).astype(int)
    
    return predictions
```

11. Implémenter une fonction artificial_neuron(X, y, learning_rate = 0.1, n_iter =100) en affichant accuracy_score et le graphique de taux d’erreur. N’oublier
pas de sauvegarder les paramètres W et b

```python
def artificial_neuron(X, y, learning_rate, n_iter):
    """Entraîner un neurone artificiel avec descente de gradient."""
    # Initialiser les paramètres
    W, b = initialisation(X)

    errors = []  # Liste pour stocker le taux d'erreur
    accuracy_history = []  # Liste pour stocker l'accuracy

    for i in range(n_iter):
        # Calculer les prédictions
        z = model(X, W, b)
        A = sigmoid(z)

        # Calculer les gradients
        dW, db = Gradient(A, X, y)

        # Mettre à jour les paramètres
        W, b = update(dW, db, W, b, learning_rate)

        # Calculer les prédictions finales
        predictions = predict(X, W, b)

        # Calculer l'accuracy
        accuracy = accuracy_score(y, predictions)
        accuracy_history.append(accuracy)

        # Calculer le taux d'erreur et l'ajouter à la liste
        error_rate = 1 - accuracy
        errors.append(error_rate)

        # Afficher l'accuracy à chaque itération
        print(f"Iteration {i+1}/{n_iter}, Accuracy: {accuracy:.4f}, Erreur :{error_rate:.4f}")

    # Sauvegarder les paramètres W et b
    np.save('W.npy', W)
    np.save('b.npy', b)
     # Afficher les résultats finaux
    print("Valeur finale des paramètres :")
    print("W (Poids) : ", W)
    print("b (Biais) : ", b)
    # Retourner les paramètres et l'historique de l'accuracy

    # Afficher le graphique de taux d'erreur
    plt.plot(errors)
    plt.title("Taux d'erreur pendant l'entraînement")
    plt.xlabel('Itérations')
    plt.ylabel('Taux d\'erreur')
    plt.show()
    return W, b, accuracy_history
```

```python
W, b, accuracy_history = artificial_neuron(X, y, learning_rate=0.1, n_iter=10000)
```

Au début de l'entraînement (itérations proches de 0), le taux d'erreur est très élevé, atteignant plus de 0,5. Cela est typique car le modèle commence avec des paramètres non optimisés.(W1=1,W2=1,b=0)
mais aprés centaines d'itérations, on observe une diminution rapide du taux d'erreur. Cela indique que le modèle apprend rapidement à partir des données initiales,et Après environ 2000 itérations, le taux d'erreur commence à se stabiliser.Dans Etnre 5000 et 7000 cette plage d'itérations, il y a des fluctuations légères mais perceptibles du taux d'erreur mais après cette plage il y'a une stabilisation en 0.0600.

13. Tracer la frontière de décision (np.linspace(-1, 4, 100)) qui sépare les deux
classes

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_decision_boundary(X, y, W, b):
    """Trace la frontière de décision pour un modèle de classification linéaire"""
    
    # Créer une grille de points dans l'espace des caractéristiques
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                         np.linspace(y_min, y_max, 100))
    
    # Calculer les prédictions pour chaque point sur la grille
    Z = predict(np.c_[xx.ravel(), yy.ravel()], W, b)  # Prédictions sur la grille
    Z = Z.reshape(xx.shape)  # Reformater pour correspondre à la grille
    
    # Afficher la frontière de décision
    plt.contourf(xx, yy, Z, alpha=0.75, cmap=plt.cm.Paired)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, edgecolors='k', marker='o', s=100)

# Tracer la frontière de décision
plot_decision_boundary(X, y, W, b)
plt.title("Frontière de Décision")
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

4. Évaluer le modèle

```python
def evaluate_model(X_test, y_test, W, b):
    # Calculer les prédictions
    z = model(X_test, W, b)  # Sortie du modèle
    A = sigmoid(z)  # Appliquer la fonction sigmoïde

    # Calculer la perte
    loss = log_loss(y_test, A)

    # Calculer l'accuracy
    predictions = predict(X_test, W, b)
    accuracy = accuracy_score(y_test, predictions)

    return loss, accuracy
```
